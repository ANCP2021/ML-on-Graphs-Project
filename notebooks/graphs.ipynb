```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

G.add_node("Lynx")
G.add_node("Hare")

G.add_edge("Lynx", "Hare")


plt.figure(figsize=(6, 4))
pos = nx.spring_layout(G) 
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='skyblue', font_size=15, font_weight='bold', arrowstyle='-|>', arrowsize=20)
plt.title("Food Chain: Lynx eats Hare")
plt.show()
```

```python
G_complex = nx.DiGraph()

nodes = ["Common Raccoon Dog", "Wild Boar", "Japanese Hare", "Japanese Badger", "Rodents"]
G_complex.add_nodes_from(nodes)

edges = [("Common Raccoon Dog", "Rodents"),
         ("Common Raccoon Dog", "Japanese Hare"),
         ("Wild Boar", "Rodents"),
         ("Wild Boar", "Japanese Hare"),
         ("Japanese Badger", "Rodents")]

G_complex.add_edges_from(edges)

plt.figure(figsize=(10, 8))
pos_complex_adjusted = nx.spring_layout(G_complex, k=1.2)
nx.draw(G_complex, pos_complex_adjusted, with_labels=True, node_size=3000, node_color='lightgreen', font_size=15, font_weight='bold', arrowstyle='-|>', arrowsize=20)
plt.title("Food Chain Relationships")
plt.show()
```

```python
G_marine = nx.DiGraph()

marine_species = ["Silver Hake", "Spiny Dogfish", "Longfin Inshore Squid", "Red Hake", "American Butterfish"]
G_marine.add_nodes_from(marine_species)

marine_edges = [
    ("Silver Hake", "Silver Hake"), ("Silver Hake", "Spiny Dogfish"),
    ("Silver Hake", "Longfin Inshore Squid"), ("Silver Hake", "Red Hake"),
    ("Silver Hake", "American Butterfish"),

    ("Spiny Dogfish", "Silver Hake"), ("Spiny Dogfish", "Spiny Dogfish"),
    ("Spiny Dogfish", "Longfin Inshore Squid"), ("Spiny Dogfish", "Red Hake"),
    ("Spiny Dogfish", "American Butterfish"),

    ("Longfin Inshore Squid", "Silver Hake"), ("Longfin Inshore Squid", "Longfin Inshore Squid"),

    ("Red Hake", "Silver Hake"), ("Red Hake", "Longfin Inshore Squid"),
    ("Red Hake", "Red Hake"), ("Red Hake", "American Butterfish"),

    ("American Butterfish", "Longfin Inshore Squid")
]

G_marine.add_edges_from(marine_edges)

plt.figure(figsize=(12, 10))
pos_marine = nx.spring_layout(G_marine, k=1.5) 
nx.draw(G_marine, pos_marine, with_labels=True, node_size=3000, node_color='lightblue', font_size=15, font_weight='bold', arrowstyle='-|>', arrowsize=20)
plt.title("Marine Species Food Web")
plt.show()
```
